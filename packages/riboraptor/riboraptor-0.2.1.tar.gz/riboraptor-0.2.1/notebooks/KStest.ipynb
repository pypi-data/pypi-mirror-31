## Stochastic ordering

A real random variable $A$ is less than a random variable $B$ in the "usual stochastic order" if

$x\in (-\infty ,\infty ) $:

$\Pr(A>x)\leq \Pr(B>x){\text{ for all }}x\in (-\infty ,\infty ) \implies \Pr(A\leq x) \geq \Pr(B\leq x)$

Loosely if $F_B(t) \leq F_A(t)$ with strict inequality for at least some $t$ then $B$ 'tends to be bigger' than $A$

```python
plt.plot(a, color='red', label='a')
plt.plot(b, color='blue', label='b')
plt.legend()
```

```python
D, effect_size, p, cdf_a, cdf_b = KS_test(a, b)
```

```python
plt.plot(cdf_a, color='red', label='a')
plt.plot(cdf_b, color='blue', label='b')
plt.axvline(D)
plt.legend()
```
